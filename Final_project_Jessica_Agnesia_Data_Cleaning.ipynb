```python
#Import library
import pandas as pd
from google.colab import files
```

```python
# Load Dataset
df = pd.read_csv('/content/ECOMM DATA.xlsx - Orders.csv')
df.head()
```

```python
# Display summary information about the dataset
df.info()
```

```python
# Display the number of rows and columns in the dataset
df.shape
```

```python
# Convert 'Order Date' and 'Ship Date' columns to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
# Convert 'Ship Mode', 'Segment', 'City', 'State', 'Country', 'Market', 'Region', 'Category', 'Sub-Category', 'Order Priority' coloums to Category type
Coloums = ['Ship Mode', 'Segment', 'City', 'State', 'Country', 'Market', 'Region', 'Category', 'Sub-Category', 'Order Priority']

for kolom in Coloums:
    if kolom in df.columns:
        df[kolom] = df[kolom].astype('category')
```

```python
# Info about columns and data types
df.info()
```

```python
# Summary statistics for numeric columns
df.describe()
```

```python
# Check number of missing values per column
print(df.isnull().sum())
```

```python
# Check presentase Missing Value of Postal Code
persentase_postal_code = df['Postal Code'].isnull().sum() * 100 / len(df)
persentase_postal_code
```

```python
# Drop Postal Code because Missing Value more than 20%
df = df.drop('Postal Code', axis=1)
```

```python
# Check Unique Values in every coloums
unique_values_per_column = df.apply(lambda x: x.unique())
print(unique_values_per_column)
```

```python
# Check duplicates
df.duplicated().sum()
```

```python
# Save the DataFrame to a CSV file
df.to_csv('cleaned_E-commerce_website_sales_analysis.csv', index=False)

# Download the file to your local computer
files.download('cleaned_E-commerce_website_sales_analysis.csv')
```
